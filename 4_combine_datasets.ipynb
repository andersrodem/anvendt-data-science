```python
import pandas as pd
from tabulate import tabulate

# Load datasets
engineered_features = pd.read_csv('Datasets/engineered_features.csv', parse_dates=['datetime_utc'])
supplementary_data = pd.read_csv('Datasets/supplementary_data_overview.csv', parse_dates=['datetime_utc'])

# Rename first column of supplementary_data
supplementary_data = supplementary_data.rename(columns={supplementary_data.columns[0]: 'datetime_utc'})

# Merge datasets
combined_dataset = pd.merge(engineered_features, supplementary_data, on='datetime_utc', how='left')

# Save combined dataset
combined_dataset.to_csv('combined_dataset.csv', index=False)

# Display first few rows of the combined dataset
print("\nFirst few rows of the combined dataset:")
print(tabulate(combined_dataset.head(15), headers='keys', tablefmt='pretty', showindex=False))

# Display information about the combined dataset
print("\nInformation about the combined dataset:")
buffer = []
combined_dataset.info(buf=buffer)
info_string = "".join(buffer)
print(info_string)
```
