```python
''' multivariate time series forecasting dataset'''
import pandas as pd
import matplotlib.pyplot as plt
from flaml import AutoML
from flaml.ml import sklearn_metric_loss_score

# ignore warnings
import warnings
warnings.filterwarnings("ignore")


# Load the dataset
df = pd.read_csv('Datasets/data_NO2.csv', sep=';', parse_dates=['datetime_utc'])

# Set datetime_utc as index and ensure timezone is removed
df = df.set_index('datetime_utc')
df.index = df.index.tz_localize(None)

# Resample to hourly frequency and forward fill any missing values
df = df.resample('h').ffill()

# Fill any remaining NaN values with backward fill
df = df.bfill()

# Reset index to have datetime as a column
df = df.reset_index()

# Split data into train and test
num_samples = df.shape[0]
time_horizon = 168  # Predict for the next 168 hours (1 week)
split_idx = num_samples - time_horizon
train_df = df[:split_idx]
test_df = df[split_idx:]

X_test = test_df[['datetime_utc', 'volume_demand', 'volume_production']]
y_test = test_df['spot_price']

# Initialize AutoML
automl = AutoML()

# Configure settings
settings = {
    "time_budget": 210,  # total running time in seconds
    "metric": "mae",  # primary metric
    "task": "ts_forecast",  # task type
    "log_file_name": "spot_price_forecast.log",  # flaml log file
    "eval_method": "holdout",
    "log_type": "all",
    "label": "spot_price",
    "period": time_horizon,
    "estimator_list": ['lgbm', 'xgboost', 'xgb_limitdepth', 'rf', 'extra_tree', 'prophet', 'arima', 'sarimax', 'holt-winters'],

}

print(train_df.head())

automl.fit(dataframe=train_df, **settings)

# Compute predictions of testing dataset
y_pred = automl.predict(X_test)

# Print predicted and true labels
print("Predicted labels:", y_pred)
print("True labels:", y_test)

# Compute MAE
mae = sklearn_metric_loss_score('mae', y_true=y_test, y_predict=y_pred)
print('MAE =', mae)

# Plot results
plt.figure(figsize=(12, 6))
plt.plot(X_test["datetime_utc"], y_test, label="Actual Spot Price")
plt.plot(X_test["datetime_utc"], y_pred, label="FLAML Forecast")
plt.xlabel("Date")
plt.ylabel("Spot Price")
plt.title("Spot Price Forecast")
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
